```python
# Look for job titles
job_title = pd.unique(data['JOB_TITLE_SUBGROUP'])
print(job_title)
```

```python
# Convert wage columns to numeric values
data['PREVAILING_WAGE_PER_YEAR'] = pd.to_numeric(data['PREVAILING_WAGE_PER_YEAR'])
data['PAID_WAGE_PER_YEAR'] = pd.to_numeric(data['PAID_WAGE_PER_YEAR'])
```

```python
# COMPARE BASED ON JOB TITLES
```

```python
# Calcualte average of prevailing wage per year by job titles
avg_prevailing_wage = data.groupby('JOB_TITLE_SUBGROUP')['PREVAILING_WAGE_PER_YEAR'].mean()
avg_prevailing_wage
```

```python
# Calcualte average of paid wage per year by job titles
avg_paid_wage = data.groupby('JOB_TITLE_SUBGROUP')['PAID_WAGE_PER_YEAR'].mean()
avg_paid_wage
```

```python
# Calculate the difference between prevailing wage per year and paid wage per year
wage_difference = avg_prevailing_wage - avg_paid_wage
wage_difference.sort_values()
```

```python
# COMPARE BASED ON COMPANIES
```

```python
# Look for employers names
employer_name = data['EMPLOYER_NAME'].value_counts()
print(employer_name)
```

```python
# Number of employers
data['EMPLOYER_NAME'].nunique()
```

```python
# Calculate the average paid wage and prevailing wage for each employer
wage_by_employer = data.groupby('EMPLOYER_NAME').mean()[['PREVAILING_WAGE_PER_YEAR', 'PAID_WAGE_PER_YEAR']]
wage_by_employer
```

```python
# Calculate the difference between the average paid wage and prevailing wage for each employer
wage_by_employer['DIFFERENCE'] = wage_by_employer['PAID_WAGE_PER_YEAR'] - wage_by_employer['PREVAILING_WAGE_PER_YEAR']
wage_by_employer
```

```python
# Identify employers that tend to overpay or underpay
```

```python
overpaid = wage_by_employer[wage_by_employer['DIFFERENCE'] > 0]
overpaid
```

```python
underpaid = wage_by_employer[wage_by_employer['DIFFERENCE'] < 0]
underpaid
```
